```python
# Pandas for managing datasets
import pandas as pd
```

```python
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Seaborn for plotting and styling
import seaborn as sns
```

```python
# Read dataset
df = pd.read_csv('TV_Shows.csv', index_col=0)
```

```python
# Display first 5 observations
df.head()
```

```python
# Recommended way
sns.lmplot(x='Netflix', y='Prime Video', data=df)
 
# Alternative way
# sns.lmplot(x=df.Attack, y=df.Defense)
```

```python
# Scatterplot arguments
sns.lmplot(x='Netflix', y='Prime Video', data=df,
           fit_reg=False, # No regression line
           hue='Age')   # Color by evolution stage
```

```python
# Plot using Seaborn
sns.lmplot(x='Netflix', y='Prime Video', data=df,
           fit_reg=False, 
           hue='Age')
 
# Tweak using Matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
# Boxplot
sns.boxplot(data=df)
```

```python
# Pre-format DataFrame
stats_df = df.drop(['Title', 'typeTvShow', 'Age', 'Hulu', 'HBO Max', 'IMDb', 'Year'], axis=1)
 
# New boxplot using stats_df
sns.boxplot(data=stats_df)
```

```python
# Set theme
sns.set_style('whitegrid')
 
# Violin plot
sns.violinplot(x='typeTvShow', y='Netflix', data=df)
```

```python
tvShow_Title_colors = ['#78C850',  # Action
                    '#F08030',  # Comedy
                    '#6890F0',  # Drama
                    '#A8B820',  # Fantasy
                    '#A8A878',  # Horror
                    '#A040A0',  # Mystery
                    '#F8D030',  # Romance
                    '#E0C068',  # Thriller
                    '#EE99AC',  # Western
                   ]
```

```python
# Violin plot with Tv Shows on OTT Platforms color palette
sns.violinplot(x='typeTvShow', y='Netflix', data=df, 
               palette=tvShow_Title_colors) # Set color palette
```

```python
# Swarm plot with Tv Shows on OTT Platforms mon color palette
sns.swarmplot(x='typeTvShow', y='Netflix', data=df, 
              palette=tvShow_Title_colors)
```

```python
# Set figure size with matplotlib
plt.figure(figsize=(10,6))
 
# Create plot
sns.violinplot(x='typeTvShow',
               y='Netflix', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               palette=tvShow_Title_colors)
 
sns.swarmplot(x='typeTvShow', 
              y='Netflix', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
 
# Set title with matplotlib
plt.title('Netflix by Type TV Show')
```

```python
# Melt DataFrame
melted_df = pd.melt(stats_df, 
                    id_vars=["Netflix", "Prime Video"], # Variables to keep
                    var_name="IMDb") # Name of melted variable
melted_df.head()
```

```python
# Calculate correlations
corr = stats_df.corr()
 
# Heatmap
sns.heatmap(corr)
```

```python
# Distribution Plot (a.k.a. Histogram)
sns.distplot(df.Netflix)
```

```python
# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='Age', data=df, palette=tvShow_Title_colors)
 
# Rotate x-labels
plt.xticks(rotation=-45)
```

```python
# Factor Plot
g = sns.factorplot(x='typeTvShow', 
                   y='Netflix', 
                   data=df, 
                   hue='Age',  # Color by stage
                   col='Age',  # Separate by stage
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
 
# Doesn't work because only rotates last plot
# plt.xticks(rotation=-45)
```

```python
# Joint Distribution Plot
sns.jointplot(x='Netflix', y='Prime Video', data=df)
```
